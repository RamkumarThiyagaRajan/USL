# Data Preparation

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
df=data.drop('target',1)
```

```python
df.head(5)
```

```python
for i in df.columns:
    sb.boxplot(df[i])
    plt.show()
```

```python
for i in df.columns:
    iqr=df[i].quantile(0.75)-df[i].quantile(0.25)
    upper_bound=df[i].quantile(0.75)+1.5*iqr
    lower_bound=df[i].quantile(0.25)-1.5*iqr
    uc=df[i].quantile(0.99)
    lc=df[i].quantile(0.1)
    for j in df[i].index:
        if df.loc[j,i]>upper_bound:    
            df.loc[j,i]=uc
        if df.loc[j,i]<lower_bound:
            df.loc[j,i]=lc
            
```

# ●	Perform Kmeans, Agglomerative clustering, PCA, build a classification model the clusters.

# K-MEANS

```python
sc=StandardScaler()
df_sc=sc.fit_transform(df)
df_sc=pd.DataFrame(df_sc,columns=df.columns)
df_sc.head()
```

```python
from sklearn.cluster import KMeans

err=[]
for i in range(1,20):
    km=KMeans(n_clusters=i,random_state=10)
    km.fit(df_sc)
    err.append(km.inertia_)
    
```

```python
err
```

```python
#plotting elbow plot

plt.plot(range(1,20),err,marker='o')
plt.xlabel('no of clusters')
plt.ylabel('inertia')
plt.show()
```

```python
model=KMeans(n_clusters=3,random_state=100)
model.fit(df_sc)
```

```python
model.labels_
```

```python
df['labels']=model.labels_
```

```python
df.head()
```

```python
sb.pairplot(df,hue='labels')
```

```python
x=df.drop('labels',1)
y=df['labels']
```

```python
sc=StandardScaler()
x_sc=sc.fit_transform(x)
x_sc=pd.DataFrame(x_sc,columns=x.columns)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_sc, y, random_state = 100, 
                                                    test_size = 0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred_dt=dt.predict(x_test)
print("Accuracy:",accuracy_score(y_test, y_pred_dt))
```

# PCA

```python
data.head()
```

```python
x=data.drop('target',1)
y=data['target']
```

```python
sc=StandardScaler()
x_std=sc.fit_transform(x)
```

```python
#COVARIANCE
cov=np.cov(x_sc.T)
cov.shape
```

```python
eig_val,eig_vector=np.linalg.eig(cov)
```

```python
print(eig_val,',',eig_vector)
```

```python
variance=[i/sum(eig_val)*100 for i in sorted(eig_val,reverse=True)]
```

```python
cum_var_exp=np.cumsum(variance)
cum_var_exp
```

```python
#SCREE PLOT
plt.bar(range(13),variance)
plt.step(range(13),cum_var_exp)
plt.axhline(96,c='r')
plt.show()
```

```python
from sklearn.decomposition import PCA
mypca=PCA(n_components=10)
pca2=mypca.fit_transform(x_std)
pca_df=pd.DataFrame(pca2)
pca_df
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(pca_df, y, random_state = 100, 
                                                    test_size = 0.2)
```

```python
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred_pca=dt.predict(x_test)
print("Accuracy:",accuracy_score(y_test, y_pred_dt))
```
