# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

# Reading Datasets

```python
data=pd.read_csv('pima-indians-diabetes.data',header=None)
```

```python
data.head()
```

```python
cols={0:'pregnancies',1:'glucose',2:'blood pressure',3:'skinthickness',4:'insulin',5:'BMI',6:'diabetes',7:'age',8:'outcome'}
```

```python
data.rename(columns=cols,inplace=True)
```

```python
data.head()
```

```python
df=data.drop('outcome',1)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
for i in df.columns:
    sb.boxplot(df[i])
    plt.show()
```

```python
for i in df.columns:
    iqr=df[i].quantile(0.75)-df[i].quantile(0.25)
    upper_bound=df[i].quantile(0.75)+1.5*iqr
    lower_bound=df[i].quantile(0.25)-1.5*iqr
    uc=df[i].quantile(0.99)
    lc=df[i].quantile(0.1)
    for j in df[i].index:
        if df.loc[j,i]>upper_bound:    
            df.loc[j,i]=uc
        if df.loc[j,i]<lower_bound:
            df.loc[j,i]=lc
            
```

```python
for i in df.columns:
    sb.boxplot(df[i])
    plt.show()
```

```python
#scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
df_sc=sc.fit_transform(df)
#df_sc=pd.DataFrame(df_sc,columns=df.columns)
```

# Principle Component Analysis

```python
#COVARIANCE
cov=np.cov(df_sc.T)
cov
```

```python
eig_val,eig_vector=np.linalg.eig(cov)
```

```python
print(eig_val,',',eig_vector)
```

```python
#var_exp=eig_val/sum(eig_val)*100
#var_exp
```

```python
variance=[i/sum(eig_val)*100 for i in sorted(eig_val,reverse=True)]
variance
```

```python
cum_var_exp=np.cumsum(variance)
cum_var_exp
```

```python
#SCREE PLOT
plt.bar(range(8),variance)
plt.step(range(8),cum_var_exp)
plt.show()
```

```python
from sklearn.decomposition import PCA
mypca=PCA(n_components=7)
pca2=mypca.fit_transform(df_sc)
```

```python
pca_df=pd.DataFrame(pca2)
pca_df
```

```python
pca_df_final=pd.concat([data['pregnancies'],pca_df,data['outcome']],1)
pca_df_final
```

```python
sb.pairplot(pca_df_final,hue='outcome')
```
